```python
# The purpose of the analysis is to demonstrate regularities in fluctuations of populations of the following species:

#- hare,
#- lynx,
#- carrot.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data.txt", skiprows=1)
years = data[:,0].astype(int)
```

```python
plt.plot(years, data[:,1], label="hare")
plt.plot(years, data[:,2], label="lynx")
plt.plot(years, data[:,3], label="carrot")
plt.xticks(years[0::5])
plt.xlabel("Species")
plt.ylabel("Population")
plt.legend();
```
